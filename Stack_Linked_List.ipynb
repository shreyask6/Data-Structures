```python
class Node:
    def __init__(self, data):
        self.next = None
        self.data = data

class Stack:
    def __init__(self):
        self.top = None

    def push(self):
        data = input("Enter data to be pushed: ")
        new_node = Node(data)
        if self.top is None:
            self.top = new_node
            self.top.next = None
        else:
            new_node.next = self.top
            self.top = new_node

    def pop(self):
        if self.top is None:
            print("Stack is empty")
        else:
            if self.top.next is None:
                print("Popped Element is : ", self.top.data)
                self.top = None
            else:
                if self.top is not None:
                    print("Popped Element is : ", self.top.data)
                    self.top = self.top.next

    def display(self):
        if self.top is None:
            print("Stack is empty")
        else:
            print("Elements of stack are:")
            temp = self.top
            while temp:
                print(temp.data)
                temp = temp.next
            print("Top of stack:", self.top.data)

s = Stack()
while True:
    stack_input = int(input("\n\n1. Push\n2. Pop \n3. Display\nEnter an option:"))
    if stack_input == 1:
        s.push()
    elif stack_input == 2:
        s.pop()
    elif stack_input == 3:
        s.display()
    else:
      break
```
