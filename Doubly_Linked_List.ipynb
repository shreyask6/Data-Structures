```python
class node:
  def __init__(self,data):
    self.data = data
    self.prev = None
    self.next = None
class DoublyLL:
  def __init__(self):
    self.head = None
  def display(self):
    cur = self.head
    while cur != None:
      print(cur.data)
      cur = cur.next
  def rdisplay(self):
    cur = self.head
    while cur.next != None:
      cur = cur.next
    while cur != None:
      print(cur.data)
      cur = cur.prev
  def addbegin(self,data):
    new = node(data)
    if self.head == None:
      print("Doubly Linked List is empty adding as first element")
      self.head = new
    else:
      self.head.prev = new
      new.next = self.head
      self.head = new
  def addend(self,data):
    new = node(data)
    if self.head == None:
      print("Doubly Linked List is empty adding as first element")
      self.head = new
    else:
      cur = self.head
      while cur.next != None:
        cur = cur.next
      cur.next = new
      new.prev = cur
  def delbegin(self):
    if self.head == None:
      print("Doubly Linked List is empty")
    else:
      self.head = self.head.next
      if self.head != None:
        self.head.prev = None
  def delend(self):
    if self.head == None:
      print("Doubly Linked List is empty")
    else:
      cur = self.head
      while cur.next != None:
        cur = cur.next
      if cur.prev != None:
        cur.prev.next = None
      else:
        print("head is only element deleting head")
        self.head = None
  def delete_by_value(self, value):
      if self.head is None:
          print("Doubly Linked List is empty")
          return
      cur = self.head
      if cur.data == value:
          self.head = cur.next
          if self.head is not None:
              self.head.prev = None
          return
      while cur is not None and cur.data != value:
          cur = cur.next
      if cur is not None:
          if cur.prev is not None:
              cur.prev.next = cur.next
          if cur.next is not None:
              cur.next.prev = cur.prev
      else:
          print(f"Node with value {value} not found in the linked list")

  def addbyindex(self,data,index):
    new = node(data)
    pos = 0
    cur = self.head
    if pos == index:
      print("LL Empty, inserting as first node")
      self.head = new
    else:
      while cur != None and pos+1 != index:
        pos += 1
        cur = cur.next
      if cur != None:
        if cur.next != None:
          cur.next.prev = new
        new.next = cur.next
        new.prev = cur
        cur.next = new
      else:
        print("Index Not present")

dllist = DoublyLL()
dllist.addbegin(1)
dllist.addend(2)
dllist.addend(3)

print("Forward Display:")
dllist.display()
dllist.delete_by_value(2)

print("\nAfter Deletion by Value:")
dllist.display()
dllist = DoublyLL()
dllist.addbegin(1)
dllist.addend(2)
dllist.addend(3)

print("Forward Display:")
dllist.display()

print("\nReverse Display:")
dllist.rdisplay()
dllist.delbegin()
dllist.delend()

print("\nAfter Deletion:")
dllist.display()
```
